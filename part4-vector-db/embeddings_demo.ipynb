```python
from sentence_transformers import SentenceTransformer, util
import seaborn as sns
import matplotlib.pyplot as plt
import torch

# 1. Define 10 sentences across 3 topics: Cricket, Cooking, and Cybersecurity
sentences = [
    # Cricket
    "The batsman hit a magnificent six over the long-on boundary.",
    "The umpire signaled a wide ball after it passed outside the crease.",
    "Test cricket requires immense patience and technical skill from players.",
    
    # Cooking
    "Slow-cooking the beef in red wine makes it incredibly tender.",
    "Add a pinch of sea salt and freshly cracked black pepper for seasoning.",
    "Professional chefs emphasize the importance of high-quality knives.",
    "The recipe calls for simmering the sauce until it reduces by half.",
    
    # Cybersecurity
    "Two-factor authentication adds a critical layer of security to online accounts.",
    "The latest malware uses zero-day vulnerabilities to infect local networks.",
    "Encrypting sensitive data ensures that it remains unreadable to hackers."
]

# 2. Load the model and generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences, convert_to_tensor=True)

print(f"Generated embeddings with shape: {embeddings.shape}")
```

```python
# 3. Compute and display a 10x10 cosine similarity matrix as a heatmap
cosine_scores = util.cos_sim(embeddings, embeddings).cpu().numpy()

plt.figure(figsize=(10, 8))
sns.heatmap(cosine_scores, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("10x10 Cosine Similarity Matrix Heatmap")
plt.xlabel("Sentence Index")
plt.ylabel("Sentence Index")
plt.show()
```

```python
# 4. Find the top 2 most similar sentences for a new query
query = "The bowler took three wickets in one over"
query_embedding = model.encode(query, convert_to_tensor=True)

hits = util.semantic_search(query_embedding, embeddings, top_k=2)[0]

print(f"Query: {query}\n")
print("Top 2 most similar sentences:")
for hit in hits:
    print(f"- {sentences[hit['corpus_id']]} (Score: {hit['score']:.4f})")
```
